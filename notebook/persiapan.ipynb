## Download Data Saham

```python
import yfinance as yf
import pandas as pd
import os
```

```python
KODES = ["TLKM", "INDF", "GOTO"]
START_DATE = "2022-07-10"
END_DATE = "2025-08-01"
```

```python
for kode in KODES:
  ticker = kode + ".JK"

  DATA_DIR = f"data/{kode}"
  os.makedirs(DATA_DIR, exist_ok=True)

  df = yf.download(ticker, start=START_DATE, end=END_DATE, auto_adjust=True, progress=False)
  df.to_csv(f"{DATA_DIR}/{kode}_hisotris.csv", index=True)
```

### Jumlah video

```python
file_TLKM = "TLKM/data/id_tlkm.json"

def load_used_ids():
    if os.path.exists(file_TLKM):
        with open(file_TLKM, 'r') as f:
            return set(json.load(f))
    return set()

list_id_TLKM = load_used_ids()
TLKM_id = pd.DataFrame(list_id_TLKM)
len_id_TLKM = len(TLKM_id)
```

```python
file_INDF = "INDF/data/id_indf.json"

def load_used_ids():
    if os.path.exists(file_INDF):
        with open(file_INDF, 'r') as f:
            return set(json.load(f))
    return set()

list_id_INDF = load_used_ids()
INDF_id = pd.DataFrame(list_id_INDF)
len_id_INDF = len(INDF_id)
```

```python
file_GOTO = "GOTO/data/id_goto.json"

def load_used_ids():
    if os.path.exists(file_GOTO):
        with open(file_GOTO, 'r') as f:
            return set(json.load(f))
    return set()

list_id_GOTO = load_used_ids()
GOTO_id = pd.DataFrame(list_id_GOTO)
len_id_GOTO = len(GOTO_id)
```

```python
print(f"TLKM : {len_id_TLKM}")
print(f"INDF : {len_id_INDF}")
print(f"GOTO : {len_id_GOTO}")
total_video = len_id_TLKM + len_id_INDF + len_id_GOTO
print(total_video)
```
